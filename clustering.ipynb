

> **Second part**



 Now I am using the KMeans and AgglomerativeClustering algorithms to perform clustering on the Iris dataset. 
 I determine the optimal number of clusters using the elbow method and silhouette score. Then fit KMeans and hierarchical clustering models with the optimal number of clusters and plot pair plots and scatter plots of the features with the cluster labels.

```python
#@title Default title text
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd
import matplotlib.pyplot as plt

iris = load_iris()

# create a dataframe with features and target
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# determine the optimal number of clusters using the elbow method
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method')
plt.show()

# determine the optimal number of clusters using the silhouette score
sil_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    sil_scores.append(silhouette_score(df, kmeans.labels_))

plt.figure(figsize=(10, 5))
plt.plot(range(2, 11), sil_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.title('Silhouette Method')
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans, AgglomerativeClustering
import pandas as pd

iris = load_iris()

# create a dataframe with features and target
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# fit a KMeans model with optimal number of clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df)
df['kmeans_cluster'] = kmeans.predict(df)

# fit a hierarchical model with optimal number of clusters
hierarchical = AgglomerativeClustering(n_clusters=2)
df['hierarchical_cluster'] = hierarchical.fit_predict(df.iloc[:, :-1])
```

```python
import seaborn as sns

sns.pairplot(df, hue='kmeans_cluster', palette='Set2')
sns.pairplot(df, hue='hierarchical_cluster', palette='Set2')
```



> **Third part**


Last part its time to perform principal component analysis (PCA) on the Iris dataset to reduce its dimensionality to two. Then fit KMeans and hierarchical clustering models with three clusters using the PCA data and plot scatter plots of the principal components with the cluster labels. Finally, I compute the silhouette score for KMeans and hierarchical clustering with the original data.

```python
from sklearn.decomposition import PCA

# instantiate PCA with 2 components
pca = PCA(n_components=2)

# fit and transform the data
pca_data = pca.fit_transform(df.iloc[:, :-2])

# create a new dataframe with the PCA data
pca_df = pd.DataFrame(data=pca_data, columns=['PC1', 'PC2'])

# add the cluster labels to the PCA dataframe
pca_df['kmeans_cluster'] = df['kmeans_cluster']
pca_df['hierarchical_cluster'] = df['hierarchical_cluster']

# plot the PCA data
sns.scatterplot(x='PC1', y='PC2', hue='kmeans_cluster', data=pca_df, palette='Set2')
sns.scatterplot(x='PC1', y='PC2', hue='hierarchical_cluster', data=pca_df, palette='Set2')
```

```python
# fit KMeans with 3 clusters using the PCA data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
kmeans_pca.fit(pca_data)

# add the KMeans cluster labels to the PCA dataframe
pca_df['kmeans_cluster_pca'] = kmeans_pca.labels_

# fit hierarchical clustering with 3 clusters using the PCA data
hierarchical_pca = AgglomerativeClustering(n_clusters=3)
hierarchical_pca.fit(pca_data)

# add the hierarchical cluster labels to the PCA dataframe
pca_df['hierarchical_cluster_pca'] = hierarchical_pca.labels_
```

```python
from sklearn.metrics import silhouette_score

# compute silhouette score for KMeans with original data
kmeans_score = silhouette_score(df.iloc[:, :-2], df['kmeans_cluster'])

# compute silhouette score for hierarchical clustering with original data
hierarchical_score = silhouette_score(df.iloc[:, :-2], df['hierarchical_cluster'])

# compute silhouette score for KMeans with PCA data
kmeans_pca_score = silhouette_score(pca_data, kmeans_pca.labels_)

# compute silhouette score for hierarchical clustering with PCA data
hierarchical_pca_score = silhouette_score(pca_data, hierarchical_pca.labels_)

print('Silhouette score without PCA:')
print(f'KMeans: {kmeans_score:.3f}')
print(f'Hierarchical clustering: {hierarchical_score:.3f}')
print('\nSilhouette score with PCA:')
print(f'KMeans: {kmeans_pca_score:.3f}')
print(f'Hierarchical clustering: {hierarchical_pca_score:.3f}')
```
